## All time and all terms

```python
Class = data.ix[data['catalognbr']==140]
```

```python
#Enrollment
Class['total_students'].plot()
```

```python
#gender makeup of class
gender = Class[['male','female']]
gender.plot(title='Gender Fractions vs. time')
print 'mean ='
print gender.mean()
gender.hist()
```

```python
##Grades in class
grades = Class[['mean_grade_male','mean_grade_female']]
grades[grades == 0] = NaN
grades.plot(title='Final Grade vs. Time')
plt.show()

print '*Mean Grades*: '
print grades.mean() 
```

```python
##Grade distribution over time
gmale = grades['mean_grade_male']
gfemale =  grades['mean_grade_female']
gmale.hist(bins=25)
#plt.xlim(1.5,4)

#plt.show()

gfemale.hist(bins =25)
plt.title('Male and Female Final Grade Distribution')
plt.show()

print 'Male Grade mean =',gmale.mean()
print 'Male Grade std =',gmale.std()
print 'Female Grade mean =',gfemale.mean()
print 'Female Grade std =',gfemale.std()

```

```python
#Plotting Gender Differences                
diff = gmale - gfemale
diff.plot(title='Difference in Male and female performance in class vs. time')
plt.show()

diff.hist()
print "Average Difference Across time:" ,diff.mean()
print "STD ", diff.std()

```

```python
#GPA make up
GPA = Class[['mean_cumgpa_male','mean_cumgpa_female']]
GPA.plot()
plt.show()

GPA.mean()
```

```python
#GPA distribution over time
#mgpa = data.ix[data['catalognbr']==140,'mean_cumgpa_male']
#fgpa = data.ix[data['catalognbr']==140,'mean_cumgpa_female']
mgpa = Class['mean_cumgpa_male']
fgpa = Class['mean_cumgpa_female']
mgpa.hist(bins=25)
fgpa.hist(bins=25)
plt.title('Male and Female final GPA Distribution')
plt.show()

print 'Male GPA mean =',mgpa.mean()
print 'Male GPA std =',mgpa.std()
print 'Female GPA mean =',fgpa.mean()
print 'Female GPA std =',fgpa.std()
```

```python
#Incoming GPA makeup
nGPA = Class[['mean_ncumgpa_male','mean_ncumgpa_female']]
nGPA.plot()
plt.show()

nGPA.mean()
```

```python
#GPA distribution over time
mngpa = Class['mean_ncumgpa_male']
fngpa = Class['mean_ncumgpa_female']
mgpa.hist(bins=25)
fgpa.hist(bins=25)
plt.title('Male and Female incoming GPA Distribution')
plt.show()

print 'Male GPA mean =',mngpa.mean()
print 'Male GPA std =',mngpa.std()
print 'Female GPA mean =',fngpa.mean()
print 'Female GPA std =',fngpa.std()
```
