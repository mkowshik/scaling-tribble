```python
##Iterating through the data set
length = len(problems['w1p1'])
length2 = len(problems.ix[0])
questions = []  #number of questions each student answers

for i in range(length):
    count = 0
    for j in range(length2):
        if problems.ix[i,j] > 0:
            count += 1
    
    questions.append(count)


```

```python
##Make bins

none = 0
low = 0
modlow = 0
modhigh = 0
high = 0
high1 = 0
high2 = 0
for i in range(length):
    if questions[i] == 0:
        none +=1
    if questions[i] >= 1 and questions[i] <=3:
        low +=1
    if questions[i] >= 4 and questions[i] <=6:
        modlow +=1
    if questions[i] >= 7 and questions[i] <=10:
        modhigh +=1
    if questions[i] >= 11 and questions[i] <=15:
        high +=1
    if questions[i] >= 16 and questions[i] <=20:
        high1 +=1
    if questions[i] >= 21 and questions[i] <=24:
        high2 +=1
```

```python
#name = np.array(['none','low','modlow','modhigh','high'])
numbers = (none,low,modlow,modhigh,high,high1,high2)
ind = np.arange(7)
width = 0.35
p1 = plt.bar(ind,numbers, color='r')
plt.xticks(ind+width/2., ('0', '1-3', '4-6', '7-10', '11-15','16-20','21-24') )
plt.xlabel('Questions Answered')
plt.ylabel('Number of students')
plt.title('Number of Students vs. Number of Questions answered')
plt.show()
```
